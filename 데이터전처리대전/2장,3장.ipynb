## 집약

### 파이썬에서는 dataframe에서 group_by 함수를 호출해 매개변수에 집약 단위를 설정한 후 집약함수를 호출해 구현할 수 있음

### 데이터 수를 계산하는 집약함수는 size 함수
### 유니크 카운트를 실행하는 집약함수는 nunique 함수

```python
rsv_cnt_tb = reserve_tb.groupby("hotel_id").size().reset_index()
rsv_cnt_tb
```

```python
reserve_tb.groupby("hotel_id").count()
```

```python
rsv_cnt_tb.columns = ["hotel_id", "rsv_cnt"]
rsv_cnt_tb
```

```python
cus_cnt_tb = reserve_tb.groupby(["hotel_id"])["customer_id"].nunique().reset_index()
cus_cnt_tb
```

```python
# 집약 결과의 열 이름을 설정
cus_cnt_tb.columns = ["hotel_id", "cus_cnt"]
display(cus_cnt_tb)
```

```python
# agg 함수의 매개변수에 dictionary 오브젝트를 지정하여 집약 처리를 한꺼번에 지정할 수 있음
result = reserve_tb.groupby("hotel_id").agg({
    "reserve_id" : "count",
    "customer_id" : "nunique"
})

result
```

## 합계값

```python
result = reserve_tb.groupby(["hotel_id", "people_num"])["total_price"].sum().reset_index()
result
```

```python
result.rename(columns = {
    "total_price" : "price_sum"}, inplace = True
)
```

```python
result
```

```python
import numpy as np
result = reserve_tb.groupby(["hotel_id"]).agg({"total_price" : ["max","min","mean","median",lambda x : np.percentile(x,q=20)]}).reset_index()
result
```

```python
result.columns
```

```python
result.columns = ["hotel_id", "price_max", "price_min","price_mean","price_median","price_20per"]
result
```

```python
result = reserve_tb.groupby(["hotel_id"]).agg({"total_price" : ["var","std"]}).reset_index()
result
```

```python
result.columns = ["hotel_id","price_var","price_std"]
```

```python
result
```

```python
# 데이터가 한 건이면, 분산값과 표준편차값이 NA가 되므로 0으로 변환

result.fillna(0, inplace = True)
# result.fillna(values = {"price_var" : 0, "price_std" : 0}, inplace = True)
# 특정 열에만 변환이 적용되도록 하여 의도치 않은 NA의 변환을 막을 수 있음
```

## 파이썬에는 최빈값 구하는 mode 함수


```python
# round 함수로 반올림 한 후 mode 함수로 최빈값 계산
reserve_tb["total_price"].round(-3).mode()
```

## 순위 계산하기

```python
reserve_tb["reserve_datetime"] = pd.to_datetime(\
                                               reserve_tb["reserve_datetime"], format = "%Y-%m-%d %H:%M:%S")
```

```python
reserve_tb["log_no"] = reserve_tb.groupby(["customer_id"])["reserve_datetime"].rank(ascending = True, method = "first")
# method = "first" => 동점 관측치 중 먼저 나타나는 관측치부터 순위 부여
reserve_tb
```
